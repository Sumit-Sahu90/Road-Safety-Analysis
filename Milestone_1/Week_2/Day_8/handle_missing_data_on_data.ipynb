Handling the missing values of wind speed(mph) and precipitation

```python
temp_df =   us_accidents_df[['Wind_Speed(mph)', 'Precipitation(in)']]
temp_df.head()
```

```python
from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer

# Step-1: Randomly sample 100,000 rows from the dataset
sample_df = temp_df.sample(n=100000, random_state=42)

# Step-2: Perform regression imputation on the sample using IterativeImputer
reg_imputer = IterativeImputer(random_state=42)
sample_imputed = reg_imputer.fit_transform(sample_df)
sample_imputed_df = pd.DataFrame(sample_imputed, columns=sample_df.columns)

# Step-3: This sample_imputed_df is the imputed subset
sample_imputed_df.head()
```

Handling the missing values for Wind speed(mph) and weather condition while imputing that data

```python
# Wind_Speed: use median (robust against outliers)
us_accidents_df['Wind_Speed(mph)'] = us_accidents_df['Wind_Speed(mph)'].fillna(
    us_accidents_df['Wind_Speed(mph)'].median()
)

# Weather_Condition: use mode (most frequent)
us_accidents_df['Weather_Condition'] = us_accidents_df['Weather_Condition'].fillna(
    us_accidents_df['Weather_Condition'].mode()[0]
)
print("Null values after imputation for 'Wind_Speed(mph)':", us_accidents_df['Wind_Speed(mph)'].isnull().sum())
print("Null values after imputation for 'Weather_Condition':", us_accidents_df['Weather_Condition'].isnull().sum())
```
